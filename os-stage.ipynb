```python
# os (operating system python module pour la manipulation file/folder)

import os
import glob
from matplotlib import image
import matplotlib.pyplot as plt
```

```python
def afficher_Noms_fichiers(path):
    
    files = os.listdir(path)
    
    for name in files:
        print(name)
```

```python
# exécution de la fonction et afficher tous les sous-dossiers et fichiers existant dans le répertoire caurant :

path = '.'  #le repertoire caurant 

afficher_Noms_fichiers(path)
```

```python
# avec le chemin absolue le repertoire NORMAL

path = "C:\\Users\\Zizu\\Downloads\\test\\NORMAL"

afficher_Noms_fichiers(path)
```

## afficher l'ensemble des fichiers avec une extension précise

```python
def extraire_fichiers(path):
    
    fichiers = []

    for file in glob.glob(path):
        fichiers.append(file)
    return fichiers
```

```python
# extraire dans le répertoire courant les fichiers avec extension .pdf

path = '*.pdf'

extraire_fichiers(path)
```

```python
# extraire dans le répertoire courant les fichiers avec extension .jpg

path = "*.jpg"

f = extraire_fichiers(path)
f
```

```python
# cherhcer les fichiers commancant par la lettre 'b'

path3 = "a*"

e = extraire_fichiers(path3)
e
```

```python
# lecture d'image a partir de variable retourner par notre fonction

img = image.imread(e[0])

plt.imshow(img)
```

### déplacer les fichiers commençant par 'a' vers un dossier

```python
import shutil
```

```python
def deplace_fichier(files, destination):
    
    for file in files:
        shutil.move(file, destination)
    
```
